## ***Import Libraries***

```python
from transformers import AutoModelForCausalLM, AutoProcessor
import torch
from PIL import Image
from IPython.display import display
import io
import requests
from google.colab import files
import warnings
warnings.filterwarnings('ignore')
```

## ***Import Phi-3-vision model***

```python
model_id = "microsoft/Phi-3-vision-128k-instruct"

model = AutoModelForCausalLM.from_pretrained(model_id,
                                             cache_dir="/content/my_models/phi_3_vision",
                                             device_map="cuda",
                                             trust_remote_code=True,
                                             torch_dtype="auto",
                                             _attn_implementation="eager")
```

## ***Get the Processor***

```python
processor = AutoProcessor.from_pretrained(model_id, trust_remote_code=True)
```

## ***Image Caption Generation***

```python
uploaded = files.upload()

image = None
for filename in uploaded.keys():
    image = Image.open(io.BytesIO(uploaded[filename]))

additional_text = input("Please provide any specific details you would like to include in the caption, or press 'Enter' to omit: ")
```

```python
messages = [
    {"role": "user", "content": f"<|image_1|>\nGenerate an interesting, creative, and engaging caption for social media (like Instagram or Facebook) using the following detail: {additional_text}. The caption should be relatable, attention-grabbing, and suitable for a wide audience. Include relevant emoticons."}
]
```

```python
prompt = processor.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
```

## ***Convert Inputs to Tokens***

```python
inputs = processor(prompt, [image], return_tensors="pt").to("cuda:0")
```

```python
generation_args = {
    "max_new_tokens": 500,
    "temperature": 0.0,
    "do_sample": False,
}
```

## ***Inference***

```python
generate_ids = model.generate(**inputs, eos_token_id=processor.tokenizer.eos_token_id, **generation_args)
```

```python
generate_ids = generate_ids[:, inputs['input_ids'].shape[1]:]
response = processor.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]

print(response)
```
